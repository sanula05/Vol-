```python
import random 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Genatating Intrest Rates Using Stochastic Modelling Techniques 
```

```python
def Random_rates(num_elm,num_set):
   
    t_r = []
    x = 0
    for z in range(num_set):
        r_r = []
        for i in range(num_elm):
            x = (np.random.lognormal(mean=0.0, sigma=1.0, size=None))
         

            r_r.append((x**2)**0.5)
        t_r.append(r_r)
    for row in t_r:
        plt.plot(row)
    
    plt.xlabel("Time")
    plt.ylabel("Rates")
    plt.title("Intrest Rates")
    plt.show()
    
        
    return t_r
def Avr_rate(arr):
    arr = np.array(arr)
    col_avgs = arr.mean(axis=0)
    plt.plot(col_avgs)
    plt.xlabel("Time")
    plt.ylabel("Avg Rates")



# print(Random_rates(100,5))
print(Avr_rate(Random_rates(100,1000)))
```
