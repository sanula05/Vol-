```python
import random as r 
import numpy as np
import matplotlib.pyplot as plt


def Random_rates(num_elm,num_set):
   
    t_r = []
    x = 0
    for z in range(num_set):
        r_r = [0]
        for i in range(num_elm):
            x = (r_r[i-1]+r.uniform(-0.5,0.5))
         

            r_r.append((x**2)**0.5)
        t_r.append(r_r)
    for row in t_r:
        plt.plot(row)
    
    plt.xlabel("Time")
    plt.ylabel("Rates")
    plt.title("Intrest Rates")
    plt.show()
        
    return t_r


print(Random_rates(100,5))
```
